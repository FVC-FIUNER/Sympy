# Introducción a Sympy

La idea de este documento es brindar al alumno una breve guía descriptiva de la utilización de la librería Sympy como herramienta para la resolución del Trabajo de Laboratorio Computacional n° 2.

El eje central de esta librería es el manejo de expresiones simbólicas. Para definir una expresión simbólica, primeramente debemos definir los *símbolos* que vamos a utilizar o que van a estar relacionados de cierta manera con nuestra expresión

```python
import sympy as sp
```

Por defecto, las variables simbólicas son complejas. Por tanto si deseamos definir una variable real, debemos aclararlo

```python
t = sp.symbols('t',real = True)
```

Hecho esto, estamos en condiciones de definir expresiones simbólicas. Por ejemplo $f(t) = t^2$

```python
f_t = t**2
f_t
```

Ahora podemos operar y manipular nuestra función $f(t)$. Por ejemplo, tomando su derivada

```python
f_t_prima = f_t.diff()
f_t_prima
```

Y finalmente podemos mostrar las funciones realizando un plot

```python
# Antes de realizar el plot cambiamos el formato de la imagen de salida para obtener una imagen de alta calidad
%config InlineBackend.figure_format = 'svg'

plot = sp.plot(f_t,(t,-2,2),ylim=(0,2),xlabel = 't [s]', ylabel='Amplitud',label='f(t)',line_color='red',legend=True,show=False);
plot.extend(sp.plot(f_t_prima,(t,-2,2),ylim=(0,2), line_color='blue',label='f\'(t)', show=False))
plot.show()
```
